```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def basic_information(file):
    data = pd.read_csv(file, sep = '\t')
    df_name = file.split('.')[0]
    print('Data set '+ df_name + ' shape is: ' + str(data.shape) )
    print('Numer of examples is: ' + str(data.shape[0]) )
    print('Number of attributes is :' + str(data.shape[1]) )
    print('----------------------------------------------------')
    print('Number of null values in every of columns is below')
    print(data.isnull().sum())
    print('----------------------------------------------------')
    print('Percentage of null values in every of columns is below')
    print(data.isnull().sum()/data.shape[0] * 100)
    print('----------------------------------------------------')
    print('Number of unique values in every of columns is below')
    print(data.nunique())
    print('----------------------------------------------------')
    print('Label ditribution')
    print(data['label'].value_counts())
    imbal = data['label'].sum()/ len(data['label'])
    print('Class imbalnace in data is: ' + str(imbal))
    print('----------------------------------------------------')
    
```

```python
##### Plot Distributoion of label 
def plot_label(data):
    label_count = data['label'].value_counts()
    sns.set(style="darkgrid")
    sns.barplot(label_count.index, label_count.values, alpha=0.9)
    plt.title('Frequency Distribution of Target Class')
    plt.ylabel('Number of Occurrences', fontsize=12)
    plt.xlabel('Label', fontsize=12)
    plt.show()
```

```python
### Check distribution of numerical atrributes by positive examples
def plot_numerical_distribution(data_positiv, val):
    col_names = ['hour', 'day_of_week', 'install_week']

    fig, ax = plt.subplots(len(col_names), figsize=(16,12))

    for i, col_val in enumerate(col_names):

        sns.distplot(data_positiv[col_val], hist=True, ax=ax[i])
        ax[i].set_title('Freq dist '+col_val + ' - ' + val, fontsize=10)
        ax[i].set_xlabel(col_val, fontsize=10)
        ax[i].set_ylabel('Count', fontsize=10)

    plt.tight_layout()
    plt.show()
```

```python
#### See basic info of all given datasets
for val in df_dict.values():
    basic_information(val)
```

```python
for val in df_dict.values():
    data = pd.read_csv(val, sep = '\t')
    plot_label(data)
```

```python
for val in df_dict.values():
    data = pd.read_csv(val, sep = '\t')
    data_positiv = data[data['label'] == 1]
    plot_numerical_distribution(data_positiv, val)
```
